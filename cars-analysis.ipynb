```python
import pandas as pd
```

```python
cars_file = "cars.csv"
```

```python
cars = pd.read_csv(cars_file)
##pandas.read_csv('your_csv_path.csv', sep=';')
## use if csv file is separated by semicolon
```

```python
cars.head()
```

```python
#filter out columns we dont care to look at
cars_df = cars[['Car', 'MPG', 'Horsepower', 'Weight', 'Acceleration', 'Model', 'Origin']].copy()
cars_df.head(10)
```

```python
#Drop the first column
cars_df = cars_df.drop(cars_df.index[0])
cars_df.head()
```

```python
most_MPG = cars_df['MPG'].max()
most_MPG
```

```python
common_MPG = cars_df['MPG'].mode()
common_MPG
```
