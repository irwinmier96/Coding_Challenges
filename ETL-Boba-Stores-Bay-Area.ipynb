```python
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy import create_engine
import json
```

```python
#Looking at all the Boba locations in the Bay Area
boba_csv = "bayarea_boba_spots.csv"
boba_spots = pd.read_csv(boba_csv)
del boba_spots['Unnamed: 0']
del boba_spots['id']
boba_spots = boba_spots.rename(columns={"name": "Name",
                                        "rating": "Rating",
                                        "address": "Address",
                                        "city": "City",
                                        "lat": "Lat",
                                        "long": "Long"})
boba_spots
```

```python
boba_spots.describe()
```

```python
#Looking at all the boba locations in San Francisco
san_francisco_boba = boba_spots[boba_spots["City"] == "San Francisco"];
san_francisco_boba
```

```python
#Seeing what the most popular boba stores are in the Bay Area
most_common_boba_stores = boba_spots['Name'].value_counts()
most_common_boba_stores
```

```python
#Seeing which cities have the highest amount of boba stores
cities_with_most_boba = boba_spots['City'].value_counts()
cities_with_most_boba
```

```python
#Dataframe for the total amount of stores in the Bay Area
store_count = boba_spots.loc[:, ["City", "Name"]];
num_stores = store_count.count()[0]
pd.DataFrame({"Total Stores": [num_stores]})
```

```python
#Making a new dataframe seeing which cities have the most boba places and the percentage each city has 
boba_by_city_totals = boba_spots['City'].value_counts()
boba_by_city_percents = boba_by_city_totals / num_stores * 100;
boba_by_city = pd.DataFrame({"City": boba_by_city_totals, "Percentage of Boba Stores": boba_by_city_percents})
boba_by_city.head(20)
```

```python
import pymysql
pymysql.install_as_MySQLdb()

rds_connection_string = "root:Piedmont2019@@127.0.0.1/boba_bay_area_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
#Creating database in MySQL

boba_spots.to_sql(name='boba_stores', con=engine, if_exists='append', index=False)
boba_by_city.to_sql(name='boba_by_city', con=engine, if_exists='append', index=False)
```

```python
#Checking the tables in the database
engine.table_names()
```
