```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
disc = pd.read_csv("Datasets/LGBT_Survey_Discrimination.csv")
disc
```

```python
#Questions 
questions = pd.DataFrame(disc.loc[:, ["question_code", "question_label"]]);
questions = questions["question_label"].value_counts()
questions
```

```python
#separate datasets by country and question: Compare a western European country to an Eastern European country
#Germany and Romania
#pick three questions related to maybe school
germany = disc[disc["CountryCode"] == "Germany"]
germany
```

```python
romania = disc[disc["CountryCode"] == "Romania"]
romania
```

```python
#Last 12 months questions: Use in Germany and Romania datasets
questions = ["During the last 12 months, have you personally felt discriminated against because of being L, G, B or T at work?",
             "During the last 12 months, have you personally felt discriminated against because of being L, G, B or T at a cafe, restaurant, bar or nightclub?",
             "During the last 12 months, have you personally felt discriminated against because of being L, G, B or T by school / university personnel?"]
```

```python
# Make plots out of 
picked_questions_GER = germany[germany.question_label.isin(questions)]
picked_questions_GER
```
