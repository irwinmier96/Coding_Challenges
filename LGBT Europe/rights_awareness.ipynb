```python
# Asking Specific Questions

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rights_awareness = pd.read_csv("Datasets/LGBT_Survey_RightsAwareness.csv")
rights_awareness.head()
```

```python
rights_awareness = pd.DataFrame(rights_awareness)
```

```python
#Question to ask: In the country where you live,
#is there a law that forbids discrimination against persons because of their sexual orientation when applying for a job?
question = ["In the country where you live, is there a law that forbids discrimination against persons because of their sexual orientation when applying for a job?"]

picked_question = rights_awareness[rights_awareness.question_label.isin(question)]
picked_question.head()
```

```python
austria_q = picked_question[picked_question["CountryCode"] == "Austria"]
```

```python
austria_q.groupby("subset").answer.sum()
austria_q
```
